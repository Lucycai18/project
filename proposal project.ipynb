Analysis 1:
-Research question: 
Is there an association between individuals’ confidence in their ability to resolve conflicts in relationships and their self-reported ability to remain calm when facing difficulties?

-Variables:
1.PSYCH_dang_liu_social_self_efficacy_scale_conflicts：ordinal variable from the 2022 cross survey，assessing the confidence in people's ability to resolve conflicts in relationships through 5 categorical responses.
2.PSYCH_general_self_efficacy_scale_calm：nominal variable from the 2022 cross survey，check true or false to the statements - I can remain calm when facing difficulties because I can rely on my coping abilities through binary responses.

-Visualizations：
Grouped Bar Chart. The grouped bar chart displays the "True" and "False" groups of each confidence level side by side. This chart can help me directly compare the proportion of "True" and "False" responses in each confidence level, which is convenient to observe whether there is a significant difference in the proportion of "True" or "False" as self-confidence changes.

-Analyses：I will use the data from 2022 cross survey， calculating the difference in the proportion of "True" responses across the different confidence levels.And then，under the null hypothesis of no association, randomly shuffle the calmness labels (True/False) to create multiple permutation samples. For each permutation sample, calculate the test statistic, building a distribution of these values. Finally，compare the original test statistic to the permutation distribution to compute the p-value. 

-Assumptions：
random sampling/Independence of Observations/

-Hypothesis:
Null Hypothesis (Ho):There is no association between individuals’ confidence in their ability to resolve conflicts in relationships and their self-reported ability to remain calm when facing difficulties
Alternative Hypothesis (H1):There is an association between individuals’ confidence in their ability to resolve conflicts in relationships and their self-reported ability to remain calm when facing difficulties


-Possible Results:

Significant association: p-value <= 0.05. (Reject Null Hypothesis)
No significant association: p-value > 0.05. (Fail to reject Null Hypothesis)

-Relevance to question:

Help me determine if individuals’ confidence in their ability to resolve conflicts in relationships is associated with their self-reported ability to remain calm when facing difficulties.

-Analyses：Assesse the association between work hours and emotional exhaustion by randomly shuffling data labels. Independence of Observations/Random sampling/Homogeneity of Variance

Null Hypothesis (Ho):There is no association between working hours per week and the frequency of feeling emotionally exhausted due to work

-Possible Results:

Significant association: p-value <= 0.05. (Reject Null Hypothesis)

Help me determine if the number of hours worked per week is associated with the frequency of feeling emotionally exhausted due to work
